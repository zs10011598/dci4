Se elimina la semana para recalcularla a partir de la fecha

```python
data_clean.drop(columns='SEMANA', inplace=True)
```

```python
data_clean['FECHA'] = data_clean['FECHA'].interpolate(method='pad', limit=4000)
data_clean.info()
```

```python
data_clean['FECHA'].unique()
```

```python
data_clean['FECHA'] = data_clean['FECHA'].apply(lambda x: ' '.join(str(x).split()[1:]))
```

```python
months = {
    'January':'01',
    'February':'02',
    'March':'03',
    'April':'04',
    'May':'05',
    'June':'06',
    'July':'07',
    'August':'08',
    'September':'09',
    'October':'10',
    'November':'11',
    'December':'12'
}
data_clean['month'] = data_clean['FECHA'].apply(lambda x: months[str(x).split()[0]])
```

```python
data_clean['day'] = data_clean['FECHA'].apply(lambda x: x.split()[1][:-1])
```

```python
data_clean['year'] = data_clean['FECHA'].apply(lambda x: x.split()[2])
```

```python
data_clean['fecha_chida'] = data_clean['year'] +'-'+ data_clean['month'] +'-'+ data_clean['day']
```

```python
#data_clean['fecha_chida'] = pd.to_datetime(data_clean['fecha_chida'], format='%Y-%M-%d')
data_clean['fecha_chida'] = pd.to_datetime(data_clean[['year', 'month', 'day']])
data_clean['fecha_chida'].unique()
```
