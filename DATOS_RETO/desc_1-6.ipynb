## Pregunta 2

```python
df = pd.read_csv('./datos_limpios/cln_extrusion.csv', parse_dates=['fecha_chida'])
df = df[['SEMANA', 'fecha_chida']].groupby('SEMANA').count()
df = df.rename({'fecha_chida':'freq'}, axis=1)
print(df)
```

```python
df.unstack().plot(kind='barh', figsize=(10,10))
```

```python
df = pd.read_csv('./datos_limpios/cln_impresion.csv', parse_dates=['FECHA'])
df = df[['SEMANA', 'FECHA']].groupby('SEMANA').count()
df = df.rename({'FECHA':'freq'}, axis=1)
print(df)
```

```python
df.unstack().plot(kind='barh', figsize=(10,10))
```

```python
df = pd.read_csv('./datos_limpios/cln_sellado.csv', parse_dates=['FECHA'])
df = df[['SEMANA', 'FECHA']].groupby('SEMANA').count()
df = df.rename({'FECHA':'freq'}, axis=1)
print(df)
```

```python
df.unstack().plot(kind='barh', figsize=(10,10))
```

## Pregunta 6

```python
df = pd.read_csv('./datos_limpios/cln_extrusion.csv', parse_dates=['fecha_chida'])

df.info()
```

```python
# No hay datos nulos en operador

df['OPERADOR'].unique()
```

```python
# hay 9 operadores de excrusion

df_x = df[['OPERADOR', 'fecha_chida']].groupby(['OPERADOR']).count()

df_x = df_x.rename({'fecha_chida':'turnos'}, axis=1)
```

```python
df_x.mean() # Promedio de turnos en excrusion
```
