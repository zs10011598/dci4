### Pregunta 8
**Como se distribuyen los calibres en periodos determinados por 3 meses y en periodos determinados por cada tres semanas?**

```python
# parsear tipo de dato a datetime
df_ex['fecha_chida'] = pd.to_datetime(df_ex['fecha_chida'])
df_se['FECHA'] = pd.to_datetime(df_se['FECHA'])
```

```python
# gets the sliding window element of a series
def slide_window(series, window_size, step):
    i = 0
    while (i + step + 1) < len(series):
        yield series[i:i+window_size].values
        i += step
```

```python
# Cada 3 Semanas
pedido_sem = df_ex.groupby(['week_chida']).sum()['CALIBRE PEDIDO']
pedido_sem = pd.DataFrame([x for x in slide_window(pedido_sem.index, 3, 3)])

real_sem = df_ex.groupby(['week_chida']).sum()['CALIBRE REAL']
real_sem = pd.DataFrame([x for x in slide_window(real_sem.index, 3, 3)])

calibre_sem = df_se.groupby(['SEMANA']).sum()['CALIBRE']
calibre_sem = pd.DataFrame([x for x in slide_window(calibre_sem.index, 3, 3)])

fig = plt.figure()
fig.set_size_inches(18.5, 10.5, forward=True)

plt.plot(pedido_sem.index, pedido_sem, label='pedido')
plt.plot(real_sem.index, real_sem, label='real')
plt.plot(calibre_sem.index, calibre_sem, label='despues de sellado')
plt.legend(loc='upper left')

plt.show()
# Cada 3 Meses
#pedido_mes = df_ex.groupby(['month']).sum()['CALIBRE PEDIDO']
#real_mes = df_ex.groupby(['month']).sum()['CALIBRE REAL']
#calibre_mes = df_se.groupby(['month']).sum()['CALIBRE']

```
