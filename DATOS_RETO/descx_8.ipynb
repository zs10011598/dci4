```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_ex = pd.read_csv('datos_limpios/cln_extrusion.csv')
print(df_ex.dtypes)
df_ex.describe()
```

```python
df_se = pd.read_csv('datos_limpios/cln_sellado.csv', index_col=0)
print(df_se.dtypes)
df_se.describe()
```

### Pregunta 8
**Como se distribuyen los calibres en periodos determinados por 3 meses y en periodos determinados por cada tres semanas?**

Las distribuciones a lo largo del tiempo se muestran en las siguientes gráficas.

```python
# parsear tipo de dato a datetime
df_ex['fecha_chida'] = pd.to_datetime(df_ex['fecha_chida'])
df_se['FECHA'] = pd.to_datetime(df_se['FECHA'])
```

```python
# gets the sliding window element of a series
def slide_window(series, window_size, step):
    i = 0
    while (i + step + 1) <= len(series):
        yield series[i:i+window_size].values
        i += step
```

```python
# Cada 3 Semanas
pedido_sem = df_ex.groupby(['week_chida']).sum()['CALIBRE PEDIDO']
pedido_sem3 = pd.DataFrame([sum(x) for x in slide_window(pedido_sem, 3, 3)])

real_sem = df_ex.groupby(['week_chida']).sum()['CALIBRE REAL']
real_sem3 = pd.DataFrame([sum(x) for x in slide_window(real_sem, 3, 3)])

calibre_sem = df_se.groupby(['SEMANA']).sum()['CALIBRE']
calibre_sem3 = pd.DataFrame([sum(x) for x in slide_window(calibre_sem, 3, 3)])
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5, forward=True)

plt.plot(pedido_sem3.index, pedido_sem3, label='pedido')
plt.plot(real_sem3.index, real_sem3, label='real')
plt.plot(calibre_sem3.index, calibre_sem3, label='despues de sellado')
plt.legend(loc='upper left')

plt.show()
```

```python
# cada 3 meses
pedido_mes = df_ex.groupby(['month']).sum()['CALIBRE PEDIDO']
pedido_mes3 = pd.DataFrame([sum(x) for x in slide_window(pedido_mes, 3, 3)])

real_mes = df_ex.groupby(['month']).sum()['CALIBRE REAL']
real_mes3 = pd.DataFrame([sum(x) for x in slide_window(real_mes, 3, 3)])

calibre_mes = df_se.groupby(['month']).sum()['CALIBRE']
calibre_mes3 = pd.DataFrame([sum(x) for x in slide_window(calibre_mes, 3, 3)])
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5, forward=True)

plt.plot(pedido_mes3.index, pedido_mes3, label='pedido')
plt.plot(real_mes3.index, real_mes3, label='real')
plt.plot(calibre_mes3.index, calibre_mes3, label='despues de sellado')
plt.legend(loc='upper left')

plt.show()
```
